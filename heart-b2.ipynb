```python
import pandas as pd
```

```python
kk = pd.read_csv('/kaggle/input/b2assignment/Heart.csv')
```

```python
kk.head()
```

```python
kk.columns
```

```python
kk.drop(columns= ['Unnamed: 0'],inplace=True,axis=1)
```

```python
kk.head()
```

```python
kk.isnull().sum()
```

```python
kk.describe()
```

```python
from sklearn.impute import SimpleImputer
```

```python
n = SimpleImputer(strategy = 'mean')
s = SimpleImputer(strategy = 'most_frequent')
```

```python
kk.dtypes
```

```python
number = ['Ca']
string = ['Thal']
```

```python
kk[number] = n.fit_transform(kk[number])
kk[number]
```

```python
kk[string] = s.fit_transform(kk[string])
kk[string]
```

```python
kk.isnull().sum()
```

```python
kk.head()
```

```python
kk.ChestPain.unique()
```

```python
kk.Thal.unique()
```

```python
kk['ChestPain'] = kk['ChestPain'].map({'typical':0, 'asymptomatic':1, 'nonanginal':2, 'nontypical':3})
```

```python
kk['Thal'] = kk['Thal'].map({'fixed':0, 'normal':1, 'reversable':2})
```

```python
kk.head()
```

```python
kk.AHD.unique()
```

```python
kk["AHD"] = kk['AHD'].map({'No':0, 'Yes':1})
```

```python
kk.head()
```

```python
import seaborn as sns
sns.countplot(x="Age",data = kk)
```

```python
from sklearn.model_selection import train_test_split as tts
x=kk[['Age','Sex','Ca','Thal','RestBP','ChestPain','Fbs','RestBP','Slope']]
y = kk['AHD']
```

```python
xt,xtt,yt,ytt = tts(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
```

```python
model = dtc()
```

```python
model.fit(xt,yt)
```

```python
result = model.predict(xtt)
result
```

```python
model.score(xtt,ytt)*100
```
